```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("microize/newyork-yellow-taxi-trip-data-2020-2019")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(path + "/yellow_tripdata_2020-03.csv")
```

```python
df
```

```python
#Check Number of null value
df.isnull().sum()
```

```python
df.dropna(inplace= True)
```

```python
# Check Number of Null value After Remove them
df.isnull().sum()
```

```python
# Check Number of Duplicated value
df.duplicated().sum()
```

```python
# Remove Duplicated Value
df.drop_duplicates(inplace= True)
```

```python
# Convert tpep_pickup_datetime to Date
df['Date'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Hour'] = df['Date'].dt.hour
df['Min'] = df['Date'].dt.minute
df['Sec'] = df['Date'].dt.second
```

```python
df
```

```python
# Convert tpep_dropoff_datetime to Dropoff_Date
df['Dropoff_Date'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df['Dropoff_Year'] = df['Dropoff_Date'].dt.year
df['Dropoff_Month'] = df['Dropoff_Date'].dt.month
df['Dropoff_Day'] = df['Dropoff_Date'].dt.day
df['Dropoff_Hour'] = df['Dropoff_Date'].dt.hour
df['Dropoff_Min'] = df['Dropoff_Date'].dt.minute
df['Dropoff_Sec'] = df['Dropoff_Date'].dt.second
```

```python
df
```

```python
# Convert Decimal value to int value
columns_to_convert = ['VendorID', 'passenger_count', 'RatecodeID', 'PULocationID',
                     'DOLocationID', 'payment_type', 'trip_distance','Hour', 'Min', 'Sec',
                     'Dropoff_Year', 'Dropoff_Month', 'Dropoff_Day',
                     'Dropoff_Hour', 'Dropoff_Min', 'Dropoff_Sec']

df[columns_to_convert] = df[columns_to_convert].astype(int)
```
